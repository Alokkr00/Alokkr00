```python
import pandas as pd
google_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/googleplaystore.csv')
google_data
```

```python
#What are the total number of applications in each categories?
```

```python
google_data[['App','Category']].groupby(['Category']).count()
```

```python
#What is the average rating of the applications in each categories?
```

```python
google_data[['Rating','Category']].groupby(['Category']).mean()
```

```python
#What is the average price of applications in each categories?
```

```python
google_data[['Price','Category']].groupby(['Category']).mean()
```

```python
#How many percent of applications are free and how many are paid?
```

```python
google_data['Type'].value_counts()
```

```python
a=(google_data['Type'].value_counts()['Free'])
b=(google_data['Type'].value_counts()['Paid'])
c=a+b
```

```python
print("Total number of free apps are "+ str(a))
print("Total number of paid apps are "+ str(b))
```

```python
print("Percentage of free apps are "+ str(a*100/c))
print("Percentage of paid apps are "+ str(b*100/c))
```

```python
#Which is the costliest application in the play store?
```
